```python
#IMPORTING LIABRARIES
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
os.getcwd()
```

```python
#Fetching CSV file

df = pd.read_csv("Dataset.csv")
df.head(5)
```

```python
#Number of rows
print("Number of rows present are:",df.shape[0])
```

```python
#Number of columns
print("Number of columns present are:",df.shape[1])
```

```python
#Dropping duplicates but keeping first occurrence 
df.drop_duplicates(keep='first',inplace = True)
```

```python
#Printing only 3 rows of dataset.
df.head(3)
```

```python
#Looking for null values / missing values
df.isnull().sum()
```

```python
#As there was very small amount of missing data , it is been dropped instead of using mode.
df = df.dropna(subset = ['Cuisines'])
```

```python
#No missing values
df.isnull().sum()
```

```python
#No missing values accross all columns
df.isnull().sum().sum()
```

```python
#Datatype of the column 'Aggregate rating' , no conversion needed.
df['Aggregate rating'].dtype
```

```python
#basic statistical measures summary
df['Aggregate rating'].describe()
```

```python
#Plotting
plt.hist(df['Aggregate rating'],bins=20)
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Aggregate Rating')
plt.show()
```

```python
df['Aggregate rating'].value_counts().sort_index()
```

```python
def rating_category(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 3.5:
        return 'Good'
    elif rating >= 2.5:
        return 'Average'
    else:
        return 'Poor'

df['Rating_Category'] = df['Aggregate rating'].apply(rating_category)
df['Rating_Category'].value_counts()
```

```python
#Splits - Turns a string into a list
cus_str = df['Cuisines'].str.split(',')
```

```python
#Top 10 most common cuisines
all_cus.value_counts().head(10)
```

```python
#Plotting Correlation Between Location And Rating AND Restaurant Locations Colored by Rating
plt.figure(figsize=(8, 6))
plt.scatter(
    df['Longitude'],
    df['Latitude'],
    c=df['Aggregate rating'],
    alpha=0.6
)
plt.colorbar(label='Aggregate Rating')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Locations Colored by Rating')
plt.show()
```
